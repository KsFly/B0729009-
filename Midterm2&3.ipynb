```python
import thinkdsp as td

x= td.UncorrelatedGaussianNoise()
x= x.make_wave(duration= 1, framerate= 10000)
```

```python
x.ts, x.ys
x.ys.shape
```

```python
import matplotlib.pyplot as pl
pl.plot(x.ts, x.ys)
```

```python
x.ys.max(), x.ys.mean(), x.ys.std()
```

```python
sx= x.make_spectrum()
sx.fs, sx.hs 
sx.hs.shape
```

```python
sx.plot()
```

```python
sx.plot_power()
```

```python
Sx= sx.make_integrated_spectrum()
Sx.cs.shape
```

```python
Sx.plot_power()
```

```python
Sx.estimate_slope()
```

```python
Sx.plot_power()
pl.xscale('log')
pl.yscale('log')
pl.grid(True)
```

```python
Sx.estimate_slope()
```

```python
y= x.cumsum()
x.ys.shape, y.ys.shape
```

```python
y.plot()
```

```python
y.ys.max(), y.ys.mean(), y.ys.std()
```

```python
sy= y.make_spectrum()
sy.plot_power()
pl.xscale('log')
pl.yscale('log')
pl.grid(True)
```

```python
sy.make_integrated_spectrum().plot_power()
pl.xscale('log')
pl.yscale('log')
pl.grid(True)
```

```python
sy.make_integrated_spectrum().estimate_slope()
```

```python
aUrl= 'https://query1.finance.yahoo.com/v7/finance/download/%5EGSPC?period1=-1325635200&period2=1588550400&interval=1d&events=history'
import pandas as pd
df= pd.read_csv(aUrl)
```

```python
df
```

```python
ys= df['Close'].values
ys= ys[-5000:]
pl.plot(ys)
```

```python
yWave= td.Wave(ys, framerate= 10000)
```

```python
yWave.plot()
```

```python
yWave.make_spectrum().make_integrated_spectrum().plot_power()
pl.xscale('log')
pl.yscale('log')
```

```python
yWave.make_spectrum().make_integrated_spectrum().estimate_slope()
```
