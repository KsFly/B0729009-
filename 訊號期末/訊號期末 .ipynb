```python
import math

def iexp(n):
    return complex(math.cos(n), math.sin(n))

def dft(xs):
    n = len(xs)
    return [sum((xs[k] * iexp(-2 * math.pi * i * k / n) for k in range(n)))
            for i in range(n)]

def dftinv(xs):
    n = len(xs)
    return [sum((xs[k] * iexp(2 * math.pi * i * k / n) for k in range(n))) / n
            for i in range(n)]

if __name__ == "__main__":
    x1 = [1,0,0,0,0,0,0,0]
    x2 = [1,1,1,1,1,1,1,1]
    x3 = [1,-1,1,-1,1,-1,1,-1]
    x4 = [3,0,2,0,2,0,2,0]
    X5 = [1,1,0,0,0,0,1,1]
    X6 = [1,1,0,0,0,0,0,1]
    x7 = [1,1,0,0]
    x8 = [1,1,0,0,0,0,0,0]
    X1 = dft(x1)
    X2 = dft(x2)
    X3 = dft(x3)
    X4 = dft(x4)
    x5 = dftinv(X5)
    x6 = dftinv(X6)
    X7 = dft(x7)
    X8 = dft(x8)
    
    print('(a)') 
    print(X1)
    print('(b)') 
    print(X2)
    print('(c)') 
    print(X3)
    print('(d)')  
    print(X4)
    print('(e)')  
    print('(f)') 
    print(x5)
    print(x6) 
    print('都不是')
    print('(g)') 
    print(X7)
    print(X8) 
    pass
```
